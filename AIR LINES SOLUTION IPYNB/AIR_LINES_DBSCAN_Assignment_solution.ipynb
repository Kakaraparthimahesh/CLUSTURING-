```python
Airline_data2_new =Airline_data2.iloc[:,1:5]
Airline_data2_new.values
```

standazation

```python
from sklearn.preprocessing import StandardScaler
stscaler = StandardScaler().fit(Airline_data2_new.values)
X = stscaler.transform(Airline_data2_new.values)
X
```

importing DBSCAN

```python
from sklearn.cluster import DBSCAN
DBSCAN()
dbscan = DBSCAN(eps=3, min_samples=8)
dbscan.fit(X)
X
```

#Noisy samples are given the label -1.

```python
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
cl
```

```python
cl['cluster'].value_counts()
```

```python
clustered0 = pd.concat([pd.DataFrame(X),cl['cluster']],axis=1)
clustered0
```
