# LabelEncoder

```python
from sklearn import preprocessing 
lable_encoder = preprocessing.LabelEncoder()
Crime_data2["Unnamed: 0"]=lable_encoder.fit_transform(Crime_data2["Unnamed: 0"])
```

# Standazation 

```python
from sklearn.preprocessing import StandardScaler
stscaler = StandardScaler().fit(Crime_data_new.values)
X = stscaler.transform(Crime_data_new.values)
X
```

# importing DBSCAN

```python
from sklearn.cluster import DBSCAN
DBSCAN()
dbscan = DBSCAN(eps=2, min_samples=5)
dbscan.fit(X)
```

#Noisy samples are given the label -1.

```python
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
cl
```

```python
cl['cluster'].value_counts()
```

```python
clustered = pd.concat([Crime_data2,cl],axis=1)       # new data set as cluster and i 
                                                     # would like to remove the noice data points 
```
